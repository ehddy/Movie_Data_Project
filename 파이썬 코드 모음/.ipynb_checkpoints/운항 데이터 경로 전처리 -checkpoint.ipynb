```python
dic = {

'GMP':'서울특별시',
    'CJU':'제주시',
    'PUS': '부산광역시'
}
```

```python
dic
```

```python
data['출발공항 주소'] = data['출발공항'].apply(lambda x: dic[x])
data['도착공항 주소'] = data['도착공항'].apply(lambda x: dic[x])


data['경로'] = data['출발공항 주소'] + '-' + data['도착공항 주소']
```
