```python
from pykrx import stock
```

```python
data = stock.get_market_ohlcv_by_date('20180101', '20220630', '079160')
```

```python
data
```

```python
data = data.reset_index()
```

```python
data['그룹'] = data['날짜'].apply(lambda x: str(x.year) + '/' +  str(x.month))
```

```python
data
```

```python
group_data = data.groupby('그룹').mean()
```

```python
group_data.to_excel('CGV 월별 평균 주가 데이터.xlsx')
```
