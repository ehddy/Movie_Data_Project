```python
# 구글 맵스 api를 이용해서 주소나 상영관 이름을 입력하면 위도, 경도 정보 수집할 수 있는 함수 실행 
import googlemaps

gmaps_key = 'AIzaSyA8s_dFBFkQUKvrRJq95PvKkkL8dYY3FIQ'  # API설정할 때 얻은 key
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
latitude = []
longitude = []
address = []
for i in range(variety_movie_data.shape[0]):
    name = variety_movie_data['주소'][i]
    geocode_result = gmaps.geocode((name), language='ko') # 
    try:
        lat  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
        lon = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
        re = geocode_result[0]['formatted_address']
        print("{0}. name: {1} lat: {2}, log: {3}, address: {4}".format(i, name, lat, lon, re))
        latitude.append(lat)
        longitude.append(lon)
        address.append(re)
    except:
        latitude.append(0)
        longitude.append(0)
        address.append(0)
        print("{0}: {1} 실패".format(i, name))

```

```python
# 45행 제외하고 모두 성공, 45행은 따로 경도 위도를 찾아주자(주소가 아니라 해당 상영관 명을 입력) 
variety_movie_data['위도'] = latitude
variety_movie_data['경도'] = longitude
```

```python
# 수집 실패한 행 번호를 찾아보자 
cinema_info_data['위도'] = latitude
cinema_info_data['경도'] = longitude
```

```python
nan_index = cinema_info_data[cinema_info_data['위도'] == 0].index.values
nan_index
```

```python
# 독립영화관뿐만 아니라, 다른 영화관도 위도 경도 데이터 수집 
latitude = []
longitude = []
address = []
for i in nan_index:
    name = cinema_info_data['영화상영관명'][i]
    geocode_result = gmaps.geocode((name), language='ko') # 
    lat  = geocode_result[0]["geometry"]["location"]["lat"] # 리스트에서 위도 추출
    lon = geocode_result[0]["geometry"]["location"]["lng"] # 리스트에서 경도 추출
    re = geocode_result[0]['formatted_address']
    print("{0}. name: {1} lat: {2}, log: {3}, address: {4}".format(i, name, lat, lon, re))
    latitude.append(lat)
    longitude.append(lon)
    address.append(re)

print('수집 완료')
```

```python
# 비어있는 인덱스에 위도, 경도 데이터를 삽입 
for idx, i in enumerate(nan_index):
    cinema_info_data.loc[i, '위도'] = latitude[idx]
    cinema_info_data.loc[i, '경도'] = latitude[idx]
```
