```python
import pandas as pd 
import datetime
import requests 
import json
from datetime import date, timedelta
import time
```

## 1. 일별 박스오피스 순위

###      1) 다양성 영화, 상업 영화 전부 포함

```python
key = "b70da09aaa9ad76068d1154144860ac5"
date_now = date.today() - timedelta(1)
date_now = date_now.strftime('%Y%m%d')
url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={}&targetDt={}'.format(key, date_now)

res = requests.get(url)

text = res.text
d = json.loads(text)
movie_list = []

for b in d['boxOfficeResult']['dailyBoxOfficeList']:
    movie_list.append([b['rank'], b['movieCd'], b['movieNm'], b['rankOldAndNew'], b['openDt'], 
    b['salesAmt'], b['salesAcc'], b['audiCnt'], b['audiAcc'], b['scrnCnt'], b['showCnt']])
    

daily_data = pd.DataFrame(data=movie_list, columns=['순위', '영화코드', '영화명', '신규진입여부', '개봉일자', '당일 매출액', '누적 매출액', '당일 관객수', '누적 관객수', '당일 상영 스크린 수', 
                                            '당일 상영 횟수'])
```

```python
# 당일(하루 전까지만 가능) 박스오피스 순위(옵션 다양: http://www.kobis.or.kr/kobisopenapi/homepg/apiservice/searchServiceInfo.do)
print('검색 날짜 : {}'.format(date_now))
daily_data
```

```python
# json 파일로 추출하고 싶다면 실행 
daily_data.to_json('daily_data_{}.json'.format(date_now))



# 엑셀로 저장하고 싶다면 
daily_data.to_excel('daily_data_{}.xlsx'.format(date_now))
```

## 3. 영화상세 정보 

```python
key = 'c5cf7452f29a492d33ba4edb3b66811f'
movie_code = '20223985'
url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={}&movieCd={}'.format(key,movie_code)
res = requests.get(url)
text = res.text
d = json.loads(text)


mov_ifo = d['movieInfoResult']['movieInfo']
# [mov_ifo['movieNm'], mov_ifo['prdtYear'], mov_ifo['showTm'], mov_ifo['openDt'], mov_ifo['prdtStatNm'], mov_ifo['typeNm'], mov_ifo['nations'][0]['nationNm'], [mov_ifo['genres'][i]['genreNm'] for i in range(len(mov_ifo['genres']))], [mov_ifo['directors'][i]['peopleNm'] for i in range(len(mov_ifo['directors']))], [mov_ifo['actors'][i]['peopleNm'] for i in range(len(mov_ifo['actors']))], mov_ifo['audits'][0]['watchGradeNm']]) 
```

```python
mov_ifo
```

```python
# 영화가 한개씩만 검색되는 문제가 있기 때문에 박스오피스 api와 결합 해야할 듯


key = 'c5cf7452f29a492d33ba4edb3b66811f'
movie_code = daily_data['영화코드'].values
movie_list = []

for code in movie_code:
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={}&movieCd={}'.format(key,code)
    res = requests.get(url)
    text = res.text
    d = json.loads(text)


    mov_ifo = d['movieInfoResult']['movieInfo']
    movie_list.append([mov_ifo['movieNm'], mov_ifo['prdtYear'], mov_ifo['showTm'], mov_ifo['openDt'], mov_ifo['prdtStatNm'], mov_ifo['typeNm'], mov_ifo['nations'][0]['nationNm'], [mov_ifo['genres'][i]['genreNm'] for i in range(len(mov_ifo['genres']))], [mov_ifo['directors'][i]['peopleNm'] for i in range(len(mov_ifo['directors']))], [mov_ifo['actors'][i]['peopleNm'] for i in range(len(mov_ifo['actors']))], mov_ifo['audits'][0]['watchGradeNm']]) 
movie_detail_data = pd.DataFrame(data=movie_list, columns=['영화명', '제작연도', '상영시간', '개봉연도', '제작상태', '영화유형', '제작국가명','장르명', '감독명', '배우명', '관람등급'])

```

```python
# 일별 박스오피스 영화 상세 정보 데이터 수집 -> 박스오피스 데이터프레임과 join 
movie_detail_data
```

```python
# 열 기준으로 병합 
daily_detail_data = pd.concat([daily_data, movie_detail_data], axis=1).drop(['개봉연도'], axis=1)
```

```python
daily_detail_data.head()
```

```python
# json 파일로 추출하고 싶다면 실행 
daily_detail_data.to_json('daily_detail_data_{}.json'.format(date_now))



# 엑셀로 저장하고 싶다면 
daily_detail_data.to_excel('daily_detail_data_{}.xlsx'.format(date_now))
```
